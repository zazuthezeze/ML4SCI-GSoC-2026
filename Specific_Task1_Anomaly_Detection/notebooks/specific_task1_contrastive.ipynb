# Specific Task 1 Progression Report on Deep Graph Detection with Contrastive Learning 

## Introduction

The goal of specific task 1 was to: classify quark/gluon jets using a model that learnt representations through contrastive loss an then to evaluate the performance

## What is Contrastive Learning?

Contrastive learning is a technique where the model identifies patterns by comparing data points and grouping similar ones and distancing non-similar ones. This is helpful for detecting anomalies since jets that are similar get grouped together and the ones which are not similar (the anomalies) are spread around so we can easily find them.

## How was this implemented? 

An encoder was trained to learn representations of jets by comparing different jets to each other, so for example: quarks and quarks and gluons and gluons get pushed together since they are similar, ending up close together in representation space. Jets that don't resemble either quarks or gluons would naturally land in a different region of the map, identifing them as anomalies. The encoder never knows if something is a quark, a gluon or anomaly it just pushes similar vectors together. 

After the encoder is done with placing the similar vectors near each other it's weights get saved, and a classifier is employed. The classifier is given the dimensional representation output which is the position on the map and starts learning whether that position corresponds to a quark or a gluon. Initially it will just guess but after some time it will recognise that jets near certain positions are very likely to be a certain type of jet.

# Performance Evaluation / comment

Best model first:

- 256dim + Weight Decay, 30 epochs:           AUC = 0.7760
- 256dim + Weight Decay + High Dropout, 30ep: AUC = 0.7759
- 128dim + Weight Decay, 50 epochs:           AUC = 0.7757
- 128dim + Weight Decay, 30 epochs:           AUC = 0.7745
- GNN Direct Classifier (Task 2):             AUC = 0.7673
- Baseline 128dim, no Weight Decay, 30ep:     AUC = 0.7076
- Baseline + High Dropout, 30ep:              AUC = 0.7001
- 256dim, no Weight Decay, 30ep:              AUC = 0.6833

AUC: measures how well the model seperates the two classes across all possible decision tresholds


From this it is evident that:

Weight decay had the single biggest impact on performance

All models with weight decay outperformed all models without weight decay regardless of any other settings

Increasing the latent dimension from 128 to 256 only helped when weight decay was present

The top 3 models are separated by only 0.0003 AUC (effectively identical performance)

High dropout on the classifier made no meaningful difference

The best contrastive model (0.7760) outperforms the GNN direct classifier from Task 2 (0.7673) despite never directly optimising for classification during encoder training, demonstrating that contrastive learning produces genuinely useful representations
